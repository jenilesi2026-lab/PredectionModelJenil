```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
data=pd.read_csv('/content/diabetes (1).csv')
data.head()
data.shape
data.describe()
data['Outcome'].value_counts()
data.groupby('Outcome').mean()
x=data.drop(columns='Outcome',axis=1)
y=data['Outcome']
print(x)
print(y)
scaler=StandardScaler()
scaler.fit(x)
x_std=scaler.transform(x)
print(x_std)
x=x_std
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)

print(x.shape,x_train.shape,x_test.shape)
cf = svm.SVC(kernel='linear')
cf.fit(x_train,y_train)

x_train_pred=cf.predict(x_train)
x_train_acc=accuracy_score(x_train_pred,y_train)
print(x_train_acc)
x_test_pred=cf.predict(x_test)
x_test_acc=accuracy_score(x_test_pred,y_test)
print(x_test_acc)
sample_input = [[
    4,      # Pregnancies
    143.7,  # Glucose
    77.7,   # BloodPressure
    0.0,    # SkinThickness
    45.9,   # Insulin
    38.1,   # BMI
    0.37,   # DiabetesPedigreeFunction
    46      # Age
]]
sample_input = scaler.transform(sample_input)

prediction = cf.predict(sample_input)

print(prediction)
```
